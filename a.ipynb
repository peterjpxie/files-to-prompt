```python
import pathspec
with open('.gitignore', 'r') as f:
    ignore_spec_text = f.read()
ignore_spec_lines = ignore_spec_text.splitlines()
ignore_spec_lines.append('.git/')

spec = pathspec.GitIgnoreSpec.from_lines(ignore_spec_lines)
matches = spec.match_tree('.')   
for m in matches:
    print(m)
```

```python
import os

os.chdir(os.getcwd())
for root, dirs, files in os.walk('.'):
    for file in files:
        relative_path = os.path.join(root, file)
        if not spec.match_file(relative_path):
            print(f"{relative_path}")
```
